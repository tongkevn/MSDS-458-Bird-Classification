```python
import os
import opendatasets as od
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
od.download("https://www.kaggle.com/datasets/soumendraprasad/sound-of-114-species-of-birds-till-2022")
```

```python
import os
import librosa.display
import matplotlib.pyplot as plt

input_directory = '/content/sound-of-114-species-of-birds-till-2022/Voice of Birds/Voice of Birds'
output_directory = '/content/drive/MyDrive/Northwestern MSDS/MSDS 458/458_Final_Notebooks/spectrogram_output'

for folder_name in os.listdir(input_directory):
    folder_path = os.path.join(input_directory, folder_name)
    if os.path.isdir(folder_path):
        output_folder = os.path.join(output_directory, folder_name)
        os.makedirs(output_folder, exist_ok=True)

        for file_name in os.listdir(folder_path):
            if file_name.endswith('.mp3'):
                mp3_path = os.path.join(folder_path, file_name)
                y, sr = librosa.load(mp3_path)
                plt.figure(figsize=(10, 4))
                spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
                librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max))
                plt.axis('off')
                output_file_path = os.path.join(output_folder, file_name.replace('.mp3', '.png'))
                plt.savefig(output_file_path, bbox_inches='tight', pad_inches=0)
                plt.close()
```
